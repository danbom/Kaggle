```python
train = pd.read_csv("/kaggle/input/DontGetKicked/training.csv")
test = pd.read_csv("/kaggle/input/DontGetKicked/test.csv")
display(train,test)
```

```python
alldata = pd.concat([train, test])
alldata
```

```python
alldata2 = alldata.drop(columns = ["IsBadBuy", "RefId", "PurchDate"])
alldata2
```

```python
alldata2["Auction"].unique()
```

```python
alldata2["Auction"] = alldata2["Auction"].replace({"ADESA" :0, "OTHER" :1, "MANHEIM" :2})
alldata2
```

```python
alldata["Make"].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
alldata2["Make"] = le.fit_transform(alldata2["Make"])
alldata2
```

```python
le.classes_
```

```python
alldata2.dtypes == object
```

```python
c = alldata2.columns[alldata2.dtypes == object]
c
```

```python
for i in c:
    print(i)
```

```python
for i in c:
    alldata2[i] = le.fit_transform(alldata2[i])
    
alldata2
```

```python
alldata2 = alldata2.fillna(-1)
alldata2
```

```python
# train2 = alldata2.iloc[:len(train), :]
# train2 = alldata2.iloc[:len(train)]
```

```python
train2 = alldata2[:len(train)]
test2 = alldata2[len(train):]
display(train2, test2)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_jobs = 4)
rf.fit(train2, train["IsBadBuy"])
result = rf.predict_proba(test2)
result
```
